# Pittsburgh Sleep Quality Index 

O Pittsburgh Sleep Quality Index, também chamado de **PSQI**, avalia de maneira subjetiva a qualidade do sono no último mês. Ele envolve **19 perguntas que abrange 7 componentes:**
1. Qualidade Subjetiva do Sono 
2. Latência do Sono 
3. Duração do Sono 
4. Eficiência do Sono 
5. Distúrbio do Sono 
6. Uso de Medicação para Dormir 
7. Sonolência e Disfunção Diurna

**Um resultado ≥ 5 indica qualidade de sono ruim! **

Todo o código foi desenvolvido para meu próprio estudo da linguagem Python, e melhorar a aplicação deste questionário nos pacientes. Porém qualquer um pode utilizá-lo para descobrir a qualidade do próprio sono. ***Em caso de um resultado ruim, procure seu médico!***

# Componente 3: Duração do Sono

O Componente 3 corresponde ao que foi respondido na questão 4.

5 - 6 horas = 2 /

# Resultado Final do PSQI 

O resultado final é a somatória do valor de cada Componente

*   **comp4res** = Coponente 4
*   **comp5res** = Componente 5 
*   **var7** = Componente 6 
*   **comp7res** = Componente 7
